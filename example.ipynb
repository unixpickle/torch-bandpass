```python
import matplotlib.pyplot as plt
import numpy as np
import torch as th

from torch_bandpass import Prism
```

```python
layer = Prism(64, mid_periods=(2, 8))
data = th.randn(1, 64) + th.arange(64) / 8
```

```python
repeated = data[..., None].repeat(1, 1, 3)
filtered = layer(repeated)

plt.plot(data[0].numpy())
plt.plot(filtered[0, :, 0].numpy(), '--')
plt.plot(filtered[0, :, 1].numpy(), '--')
plt.plot(filtered[0, :, 2].numpy(), '--')
plt.legend(['signal', 'low-pass', 'mid-pass', 'high-pass'])
plt.show()
```
